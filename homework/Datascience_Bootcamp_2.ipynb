```python
#1. Define two custom numpy arrays, say A and B. Generate two new numpy arrays by stacking A and B vertically and horizontally.
import numpy as np
array_a= np.array(["a","b","c","d","e","f"])
array_b= np.array(["a","l","m","n","o","p"])
array_c= np.hstack([array_a,array_b])
print(array_c)

array_d = np.vstack([array_a,array_b])
print(array_d)

array_e = array_a.reshape(-1,1)
array_f = array_b.reshape(-1,1)
array_g = np.vstack([array_e,array_f])
print(array_g)
```

```python
#2. Find common elements between A and B. [Hint : Intersection of two sets]
print(np.intersect1d(array_a,array_b))
```

```python
#3. Extract all numbers from A which are within a specific range. eg between 5 and 10. [Hint: np.where() might be useful or boolean masks]
array_a = np.arange(20)

# Define the specific range
lower_bound = 5
upper_bound = 10

condition = (array_a > lower_bound) & (array_a < upper_bound)
indices = np.where(condition)

extracted_numbers = array_a[indices]

print(extracted_numbers)
```

```python
#4. Filter the rows of iris_2d that has petallength (3rd column) > 1.5 and sepallength (1st column) < 5.0
from urllib.request import urlretrieve
import pandas as pd
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

filename = "iris.data"
urlretrieve(url, filename)


column_names = ['sepallength', 'sepalwidth', 'petallength', 'petalwidth', 'class']
iris_df = pd.read_csv(filename, header=None, names=column_names)

filtered_iris = iris_df[(iris_df['petallength'] > 1.5) & (iris_df['sepallength'] < 5.0)]

# Display the filtered DataFrame
print("\nFiltered DataFrame:")
display(filtered_iris)
```

```python
#1. From df filter the 'Manufacturer', 'Model' and 'Type' for every 20th row starting from 1st (row 0).
from urllib.request import urlretrieve
import pandas as pd

url = "https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv"
filename = "manufacture.data"
urlretrieve(url, filename)

manufacture_df = pd.read_csv(filename)

every_20th_row = manufacture_df.iloc[0::20]

columns_df = every_20th_row[['Manufacturer', 'Model', 'Type']]

print("Filtered DataFrame (every 20th row with selected columns):")
display(columns_df)
```

```python
#2. Replace missing values in Min.Price and Max.Price columns with their respective mean.

print("Missing values per column before filling:")
print(manufacture_df.isnull().sum())
print("\n")

mean_min_price = manufacture_df['Min.Price'].mean()
mean_max_price = manufacture_df['Max.Price'].mean()

manufacture_df['Min.Price'].fillna(mean_min_price, inplace=True)
manufacture_df['Max.Price'].fillna(mean_max_price, inplace=True)

print("Missing values per column after filling:")
print(manufacture_df.isnull().sum())
print("\n")

print("DataFrame after filling missing values (first 5 rows):")
display(manufacture_df.head())
```

```python
#3. How to get the rows of a dataframe with row sum > 100?
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
row_sum = df.sum(axis=1)
rows_with_sum_greater_than_100 = df[row_sum > 100]
print(rows_with_sum_greater_than_100)
```
